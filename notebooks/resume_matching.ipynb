```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = wordnet_lemmatizer = WordNetLemmatizer()
def clean_text(text):
    text = str(text)
    text = re.sub(r'[^\w\s]', '', text)
    text = text.lower()
    words = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(tokens)
```

```python
resume_df["cleaned_resume"] = resume_df["Resume_str"].apply(clean_text)
```

```python
job_df["clean_job"] = (
    job_df["combined_text"]
    .apply(clean_text)
)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)
```

```python
resume_embeddings = model.encode(
    resume_df["cleaned_resume"].tolist(),
    show_progress_bar=True
    
)
```

```python
resume_embeddings.shape
```

```python
job_embeddings = model.encode(
    job_df["clean_job"].tolist(),
    show_progress_bar=True
)
```

```python
job_embeddings.shape
```

```python
print(resume_embeddings.shape)

print(job_embeddings.shape)
```

```python
job_df.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(
    job_embeddings,
    resume_embeddings
)
```

```python
job_index = 0

best_resume_idx = similarity_matrix[job_index].argmax()

best_score = similarity_matrix[job_index][best_resume_idx]

print("Best Resume Index:", best_resume_idx)
print("Similarity Score:", best_score)

print("\nResume Category:")
print(resume_df.iloc[best_resume_idx]["Category"])
```

```python
top5_indices = similarity_matrix[job_index].argsort()[-5:][::-1]
print("\nTop 5 Resume Indices:", top5_indices)
print("Top 5 Similarity Scores:", similarity_matrix[job_index][top5_indices])
print("\nTop 5 Resume Categories:")
for idx in top5_indices:
    print(f"Index: {idx}, Category: {resume_df.iloc[idx]['Category']}")

```

```python
def get_top_matches(job_index, top_n=5):
    score = similarity_matrix[job_index]
    top_indices = similarity_matrix[job_index].argsort()[-top_n:][::-1]
    results = []
    for idx in top_indices:
        results.append({
            "resume_index": idx,
            "category": resume_df.iloc[idx]['Category'],
            "similarity_score": score[idx]
        })
    return pd.DataFrame(results)
```
